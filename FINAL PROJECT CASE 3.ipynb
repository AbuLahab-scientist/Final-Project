## 1 Cleaning Data  & Handling Data

#### handling NaN

```python
data = data.dropna(subset=['odometer', 'mmr', 'sellingprice', 'saledate', 'color'])

#drop kolo NaN yang <1%
```

```python
# Menghitung persentase NaN per kolom
total_nan=data.isnull().sum().sort_values(ascending=False)
percent= data.isnull().sum()/data.isnull().count().sort_values(ascending=False)
missing_data=pd.concat([total_nan, percent], axis=1, keys=['Total NaN', 'percentage'])
missing_data
```

```python
df = data.copy()
```

```python
total_nan=df.isnull().sum().sort_values(ascending=False)
total_nan

#cek total baris yang NaN di kolom
```

```python
#hitung modus
mode_transmission = df['transmission'].mode()[0]
df['transmission'] = df['transmission'].fillna(mode_transmission)

# Mengisi nilai NaN di kolom 'transmission' dengan modus
```

```python
df['body'] = df['body'].fillna('unknown')

# Mengisi nilai NaN di kolom 'body' dengan kategori baru 'unkrown'
```

```python
df['make'] = df['make'].bfill()

# isi kolom NaN di make dengan Bfill
```

```python
df['model'] = df['model'].bfill()
```

```python
df.loc[df['vin'].str.contains('wbak', na=False), 'trim'] = '750li'
#mengisi nilai NaN dengan 750 li
```

```python
df['trim'] = df['trim'].bfill()
#mengisi nilai NaN dengan nilai sebelumnya
```

```python
df.loc[df['vin'].str.contains('wbak|wbaka', na=False), 'make'] = 'bmw'
```

```python
df['condition'] = df['condition'].fillna(df['condition'].mean())
#isi NaN di kolom condition dengan rata-rata
```

### DATA PREPARATION

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

# Inisialisasi scaler
scaler = StandardScaler()

# Scaling data_numeric 'sellingprice' dan 'year'
data_numeric_scaled = data_numeric.copy()
data_numeric_scaled['sellingprice'] = scaler.fit_transform(data_numeric[['sellingprice']])
data_numeric_scaled['year'] = scaler.fit_transform(data_numeric[['year']])

```

```python
# Menyimpan ke file CSV
data_numeric_scaled.to_csv('data_numeric_scaled.csv', index=False)

print("File berhasil disimpan sebagai 'data_numeric_scaled.csv'")
```

```python
# Cek ukuran dataset asli
print(data_numeric_scaled.shape)  # Output: (jumlah_baris, jumlah_kolom)

# Hanya jumlah baris
print(len(data_numeric_scaled))   # Output: jumlah_baris
```

```python
# Pisahkan data menjadi X dan Y
X = data_numeric_scaled['year']
Y = data_numeric_scaled['sellingprice']
```

```python
 #Split data menjadi training dan testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)

#cek Jumlah Baris
```

```python
# this is needed only when you are using single feature; in this case 'TV'

import numpy as np
X_train = X_train[:,np.newaxis]
X_test = X_test[:,np.newaxis]
      # np.newaxis: menambah dimensi array
```

```python
X_train = X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

## PERFORMING LINEAR REGRESSIOn

```python
from sklearn.linear_model import LinearRegression
#import linearregression from sklearn

lr= LinearRegression()
#representing LinearRegression

lr.fit(X_train, Y_train)
# fit the model

```

```python
print('*intecept:',lr.intercept_)
print('Ini adalah nilai sellingprice (y) saat variabel year (x) bernilai 0.')

print('*coefisien:',lr.coef_)
print('Jika nilai year meningkat 1 unit, maka sellingprice akan naik sekitar 0.6567 unit')
      #print koefiesien sama intecept nya
```

## Simple Linear regression
Y = c + M1 X1 + M2 X2 +....+Mn Xn
- Y = Variabel dependen (target yang diprediksi). Contoh: Penjualan barang.
- c = Intersep (intercept): Nilai Y ketika semua x1,x2,xn tidak ada
- X1 , X2, X..,Xn =(predictors) Faktor-faktor yang memengaruhi nilai Y
- M1 , M2, M.., Mn = (koefisen Regresi) Menggambarkan seberapa besar perubahan Y jika X1 bertambah 1 unit, smntara variabel lain tetap


*karena hanya ada satu variabel independen (year), sehingga persamaannya menjadi lebih sederhana:*

sellingprice = c + M1 * X
- atau
- sellingprice=−0.00029237512847383443+0.65665928 x year

## predictions

```python
Y_predict = lr.predict(X_test)
                      #membuat hasil predict pada data test
```
